### 🎯 Objetivo del cuaderno

Este notebook explora cómo las matrices transforman datos mediante operaciones lineales.  
El enfoque es **visual e intuitivo**, orientado al rol de **Data Analyst**.

```python
import numpy as np 
import matplotlib.pyplot as plt

# Cuadrado unitario: 4 puntos + vuelta al inicio para cerrar figura
cuadrado = np.array([
    [0, 1, 1, 0, 0], # x
    [0, 0, 1, 1, 0] # y
])
```

```python
# Esta funcion nos permitira reutilizar codigo para cada tipo de transformacion
def graficar_transformacion(A, nombre="Transformacion", color="blue"):
    transformado = A @ cuadrado

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(cuadrado[0], cuadrado[1], 'gray', label='Original')
    ax.fill(cuadrado[0], cuadrado[1], 'gray', alpha=0.2)

    ax.plot(transformado[0], transformado[1], color, label=nombre)
    ax.fill(transformado[0], transformado[1], color, alpha=0.3)

    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_aspect('equal')
    ax.grid(True)
    ax.legend()
    plt.title(nombre)
    plt.show()
```

### 🔁 Rotación 90°

Estoy probando cómo una matriz puede transformar una figura. En este caso, apliqué una rotación de 90 grados al cuadrado unitario.

Lo que hace la matriz es "girar" cada punto del cuadrado en sentido antihorario. La forma general se conserva (no hay escalado ni deformación), solo cambia su orientación en el espacio.

Me interesa esto porque, aunque aún no entiendo del todo qué es PCA, sé que en algún punto mis datos van a rotarse o alinearse de otra manera para que "se entiendan mejor". Así que prefiero ir desarrollando esta intuición desde ya.


```python
# Transformacion: Rotacion del cuadrado 

# Definimos la matriz de rotacion
theta = np.pi / 2 # 90 grados

A_rot = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

# Aplicamos y graficamos 
graficar_transformacion(A_rot,"Rotacion 90º","red")
```

### 📏 Escalado: stretch y squeeze

En esta celda probé una transformación que **estira** la figura en un eje y **la aplasta** en otro. Específicamente:

- Eje X: multiplicado por 2 → **stretch**
- Eje Y: multiplicado por 0.5 → **squeeze**

El resultado es un rectángulo más ancho y más bajo.

```python
# Definimos la matriz de escalado:
A_scl = np.array([
    [2, 0], # Estira el eje x al doble
    [0, 0.5] # Aplasta el eje y a la mitad 
])

# Aplicamos y graficamos
graficar_transformacion(A_scl, "Escalado: x2 en X, x0.5 en Y", "purple")
```

### ↔️ Reflexión respecto al eje Y

Probé una matriz que hace un efecto espejo: invierte la figura respecto al eje Y.

Matemáticamente, lo que hace es multiplicar el eje X por -1, lo cual refleja todos los puntos hacia el lado opuesto del plano. El eje Y se mantiene igual.

La figura no se deforma ni rota, simplemente cambia de orientación. Es como girarla sobre un espejo vertical.


```python
# Definimos la matriz de reflexion:
A_reflex_y = np.array([
    [-1, 0],
    [0, 1]
])

# Aplicamos y graficamos
graficar_transformacion(A_reflex_y, "Reflexion respecto al eje Y", "green")
```

### 🪟 Cizalla horizontal (shear en X)

En esta transformación probé una matriz que aplica una **cizalla horizontal**, también conocida como shear.

Lo que hace es desplazar los puntos horizontalmente en función de su altura. Es decir, cuanto más alto está un punto en el eje Y, más se mueve en el eje X.

El resultado es que el cuadrado se convierte en un paralelogramo, pero sin rotarlo ni escalarlo.

Visualmente parece que alguien empujó la figura hacia un lado, como si fuera de goma.


```python
# Matriz de cizalla (horizontal):
A_shear_x = np.array([
    [1, 1], # Agrega un componente horizontal proporcional a Y 
    [0, 1]
])

# Aplicamos y graficamos
graficar_transformacion(A_shear_x, "Cizalla horizontal(X)", "orange")
```
