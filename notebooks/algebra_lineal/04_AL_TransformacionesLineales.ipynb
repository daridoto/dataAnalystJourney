# 04_AL_TransformacionesLineales.ipynb
    Objetivo: Explorar cómo las matrices aplicadas a vectores generan transformaciones lineales en el plano.
    Veremos visualmente qué hacen estas transformaciones y qué propiedades tienen.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el cuadrado: 4 vertices + volver al primero para cerrar figura
cuadrado = np.array([
    [0, 1, 1, 0, 0],
    [0, 0, 1, 1, 0]
])

# Matriz de transformacion A (por ejemplo, rotacion + escalado)
A = np.array([
    [2, 1],
    [1, 1]
])

# Aplicamos la transformacion
cuadrado_transformado = A @ cuadrado

# Graficamos
fig, ax = plt.subplots(figsize=(6,6))

# Cuadrado original
ax.plot(cuadrado[0], cuadrado[1], 'gray', label='Cuadrado original')
ax.fill(cuadrado[0], cuadrado[1], 'gray', alpha=0.2)

# Cuadrado transformado
ax.plot(cuadrado_transformado[0], cuadrado_transformado[1],'red', label='Transformado')
ax.fill(cuadrado_transformado[0], cuadrado_transformado[1], 'red', alpha=0.2)

ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_aspect('equal')
ax.grid(True)
ax.legend()
plt.title("Transformacion lineal de un cuadrado")
plt.show()
```
