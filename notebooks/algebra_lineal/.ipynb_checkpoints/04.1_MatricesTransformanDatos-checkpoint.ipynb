### üéØ Objetivo del cuaderno

Este notebook explora c√≥mo las matrices transforman datos mediante operaciones lineales.  
El enfoque es **visual e intuitivo**, orientado al rol de **Data Analyst**.

```python
import numpy as np 
import matplotlib.pyplot as plt

# Cuadrado unitario: 4 puntos + vuelta al inicio para cerrar figura
cuadrado = np.array([
    [0, 1, 1, 0, 0], # x
    [0, 0, 1, 1, 0] # y
])
```

```python
# Esta funcion nos permitira reutilizar codigo para cada tipo de transformacion
def graficar_transformacion(A, nombre="Transformacion", color="blue"):
    transformado = A @ cuadrado

    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(cuadrado[0], cuadrado[1], 'gray', label='Original')
    ax.fill(cuadrado[0], cuadrado[1], 'gray', alpha=0.2)

    ax.plot(transformado[0], transformado[1], color, label=nombre)
    ax.fill(transformado[0], transformado[1], color, alpha=0.3)

    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_aspect('equal')
    ax.grid(True)
    ax.legend()
    plt.title(nombre)
    plt.show()
```

### üîÅ Rotaci√≥n 90¬∞

Estoy probando c√≥mo una matriz puede transformar una figura. En este caso, apliqu√© una rotaci√≥n de 90 grados al cuadrado unitario.

Lo que hace la matriz es "girar" cada punto del cuadrado en sentido antihorario. La forma general se conserva (no hay escalado ni deformaci√≥n), solo cambia su orientaci√≥n en el espacio.

Me interesa esto porque, aunque a√∫n no entiendo del todo qu√© es PCA, s√© que en alg√∫n punto mis datos van a rotarse o alinearse de otra manera para que "se entiendan mejor". As√≠ que prefiero ir desarrollando esta intuici√≥n desde ya.


```python
# Transformacion: Rotacion del cuadrado 

# Definimos la matriz de rotacion
theta = np.pi / 2 # 90 grados

A_rot = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

# Aplicamos y graficamos 
graficar_transformacion(A_rot,"Rotacion 90¬∫","red")
```

### üìè Escalado: stretch y squeeze

En esta celda prob√© una transformaci√≥n que **estira** la figura en un eje y **la aplasta** en otro. Espec√≠ficamente:

- Eje X: multiplicado por 2 ‚Üí **stretch**
- Eje Y: multiplicado por 0.5 ‚Üí **squeeze**

El resultado es un rect√°ngulo m√°s ancho y m√°s bajo.

```python
# Definimos la matriz de escalado:
A_scl = np.array([
    [2, 0], # Estira el eje x al doble
    [0, 0.5] # Aplasta el eje y a la mitad 
])

# Aplicamos y graficamos
graficar_transformacion(A_scl, "Escalado: x2 en X, x0.5 en Y", "purple")
```

### ‚ÜîÔ∏è Reflexi√≥n respecto al eje Y

Prob√© una matriz que hace un efecto espejo: invierte la figura respecto al eje Y.

Matem√°ticamente, lo que hace es multiplicar el eje X por -1, lo cual refleja todos los puntos hacia el lado opuesto del plano. El eje Y se mantiene igual.

La figura no se deforma ni rota, simplemente cambia de orientaci√≥n. Es como girarla sobre un espejo vertical.


```python
# Definimos la matriz de reflexion:
A_reflex_y = np.array([
    [-1, 0],
    [0, 1]
])

# Aplicamos y graficamos
graficar_transformacion(A_reflex_y, "Reflexion respecto al eje Y", "green")
```

### ü™ü Cizalla horizontal (shear en X)

En esta transformaci√≥n prob√© una matriz que aplica una **cizalla horizontal**, tambi√©n conocida como shear.

Lo que hace es desplazar los puntos horizontalmente en funci√≥n de su altura. Es decir, cuanto m√°s alto est√° un punto en el eje Y, m√°s se mueve en el eje X.

El resultado es que el cuadrado se convierte en un paralelogramo, pero sin rotarlo ni escalarlo.

Visualmente parece que alguien empuj√≥ la figura hacia un lado, como si fuera de goma.


```python
# Matriz de cizalla (horizontal):
A_shear_x = np.array([
    [1, 1], # Agrega un componente horizontal proporcional a Y 
    [0, 1]
])

# Aplicamos y graficamos
graficar_transformacion(A_shear_x, "Cizalla horizontal(X)", "orange")
```
