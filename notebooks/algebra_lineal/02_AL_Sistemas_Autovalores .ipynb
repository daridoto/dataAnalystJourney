```python
# Resolver un sistema Ax = b

import numpy as np 

# Sistema: 2x + y = 8, x + 3y = 13
A = np.array([[2, 1], [1, 3]])
b = np.array([8, 13])

x = np.linalg.solve(A, b)

print("Solucion del sistema Ax = b:", x)
```

```python
# Autovalores: marcan direcciones invariantes, Autovectores: indican el factor de escalado

M = np.array([[2, 0], [0, 3]])

vals, vecs = np.linalg.eig(M)

print("Autovalores: ", vals)
print("Autovectores: ", vecs)
```

```python
# Grafico los autovectores para verlo
import matplotlib.pyplot as plt

origin = [0, 0]

for i in range(len(vals)):
    plt.quiver(*origin, vecs[0, i], vecs[1, i], angles='xy', scale_units='xy', scale=1, color=['r','b'][i])

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid()
plt.title("Autovectores de M")
plt.show()
```
