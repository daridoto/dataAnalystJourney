```python
# Proyectemos un vector sobre un autovector para verlo y entenderlo:

import numpy as np
import matplotlib.pyplot as plt

# Vector original
v = np.array([2, 3])

# Autovector sobre el que proyectamos
u = np.array([1, 0]) # direccion del eje x

# Calculamos la proyeccion escalar
proj_length = np.dot(v, u) / np.linalg.norm(u)

# El vector proyeccion
proj_vector = proj_length * (u / np.linalg.norm(u))

# Graficamos
origin = [0, 0]
plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(*origin, proj_vector[0], proj_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Proyeccion de v en u')

plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid()
plt.legend()
plt.title("Proyeccion de v sobre u")
plt.show()
```
